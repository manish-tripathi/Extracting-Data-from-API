### Looking at seasons best stats for 100 metres as has been described in the video

[Wikipedia Link](https://en.wikipedia.org/wiki/100_metres)

````python
dfs = pd.read_html("https://en.wikipedia.org/wiki/100_metres")
````

````python
mens100m = dfs[5]
womens100m = dfs[6]
````

````python
mens100m.head(10)
````

As part of data cleaning we will replace "[A]" and "h" that is present in "Time" column and convert it to numeric

````python
mens100m["Time"] = mens100m["Time"].str.replace("[A]","", regex= False).str.replace("h","", regex= False)
mens100m["Time"] = pd.to_numeric(mens100m["Time"])
````

````python
womens100m["Time"]
````

````python
womens100m["Time"] = womens100m["Time"] .str.replace("[A]","", regex= False).str.replace("h","", regex= False).\
str.replace("[a]","", regex= False)
womens100m["Time"] = pd.to_numeric(womens100m["Time"])
````

````python
womens100m
````

### Plot the Data

````python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
````

````python
mens100m["Name"] = mens100m["Year"].astype("str") + ": " + mens100m["Athlete"]
````

````python
womens100m["Name"] = womens100m["Year"].astype("str") + ": " + womens100m["Athlete"]
````

````python
import seaborn as sns
color_pal = sns.color_palette()
````

````python
mens100m.head()
````

````python
fig, axs =  plt.subplots(1,2, figsize = (15,15))


mens100m.sort_values("Year", ascending = False)\
.set_index("Name")["Time"].plot(kind = "barh", ax = axs[0])
ax= axs[0].set_xlim(9.4, 10.2)
ax= axs[0].set_title("Men's Season Best 100m Dash")

womens100m.sort_values("Year", ascending = False).\
set_index("Name")["Time"].plot(kind = "barh", ax = axs[1], color = color_pal[1] )
ax= axs[1].set_xlim(10.4, 11.4)
ax= axs[1].set_title("Women's Season Best 100m Dash")
plt.tight_layout()
plt.show()
````
